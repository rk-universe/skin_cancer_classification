```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
from glob import glob
from PIL import Image
# Directory with our training pictures
HAM10000_images_part_1 = os.path.join('reseze_1')
# Directory with our training pictures
HAM10000_images_part_2= os.path.join('reseze_2')
HAM10000_metadata = os.path.join('HAM10000_metadata.csv')
```

```python
data=pd.read_csv(HAM10000_metadata)
```

```python
class_counts = data['dx'].value_counts()

# Plot the class distribution
plt.figure(figsize=(8, 6))
plt.bar(class_counts.index, class_counts.values)
plt.xlabel('Class Labels')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.xticks(rotation=45)
print(class_counts)
plt.show()
```

```python
HAM10000_images_part_1_name = os.listdir(HAM10000_images_part_1)
print(HAM10000_images_part_1_name[:10])
HAM10000_images_part_2_name = os.listdir(HAM10000_images_part_2)
print(HAM10000_images_part_2_name[:10])
```

```python
for i in range(0,len(HAM10000_images_part_1_name)):
    HAM10000_images_part_1_name[i] =os.path.join(HAM10000_images_part_1,HAM10000_images_part_1_name[i])
```

```python
image = img = plt.imread(HAM10000_images_part_1_name[3])
plt.imshow(img)
```

```python
for i in range(0,len(HAM10000_images_part_2_name)):
    HAM10000_images_part_2_name[i] =os.path.join(HAM10000_images_part_2,HAM10000_images_part_2_name[i])
```

```python
HAM10000_images_part_merged = HAM10000_images_part_1_name + HAM10000_images_part_2_name
```

```python
sorted_df = data.sort_values('image_id')
```

```python
image_mapping = dict(zip(sorted_df['image_id'], HAM10000_images_part_merged))
```

```python
sorted_df['path'] = sorted_df['image_id'].map(image_mapping)
```

```python
sorted_df['cell_type_idx'] = pd.Categorical(sorted_df['dx']).codes
```

```python
desired_rows_per_class = 500

grouped = sorted_df.groupby('cell_type_idx')

balanced_df = pd.DataFrame()

# Iterate over each group
for group, data in grouped:
    # Sample the desired number of rows from each group
    sampled_data = data.sample(n=desired_rows_per_class,replace= True, random_state=42)
    # Append the sampled data to the balanced DataFrame
    balanced_df = balanced_df.append(sampled_data)

# Reset the index of the balanced DataFrame
balanced_df = balanced_df.reset_index(drop=True)
```

```python
balanced_df.head()
```

```python
balanced_df['image'] = balanced_df['path'].map(lambda x: np.asarray(Image.open(x)))
```

```python
balanced_df['image']=balanced_df['image']/255
```

```python
X_sample_i = np.asarray(balanced_df['image'].tolist())
```

```python
y_sample = balanced_df['cell_type_idx']
```

```python
y_sample = pd.get_dummies(y_sample)
```

```python
X_sample_norm = (X_sample_i - np.mean(X_sample_i))/np.std(X_sample_i)
```

```python
from sklearn.model_selection import train_test_split
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_sample_norm, y_sample,stratify=y_sample, test_size = 0.2)
```

```python
X_train_s.shape,X_test_s.shape,y_train_s.shape,y_test_s.shape
```

```python
y_test_s
```

```python
import tensorflow as tf
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(30,(5,5),strides=(1,1),padding='valid',activation='relu',input_shape=(94,94,3)))
model.add(tf.keras.layers.Conv2D(30,(3,3),strides=(1,1),padding='valid',activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=None,padding='valid'))
model.add(tf.keras.layers.Conv2D(20,(3,3),strides=(1,1),padding='valid',activation='relu'))
model.add(tf.keras.layers.Conv2D(15,(3,3),strides=(1,1),padding='valid',activation='relu'))
model.add(tf.keras.layers.Conv2D(15,(3,3),strides=(1,1),padding='valid',activation='relu'))
model.add(tf.keras.layers.GroupNormalization(groups=3))
model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=None,padding='valid'))
model.add(tf.keras.layers.Conv2D(10,(3,3),strides=(1,1),padding='valid',activation='relu'))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Normalization())
model.add(tf.keras.layers.Dense(256,activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.1))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(7,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer="Adam",loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(
    X_train_s, y_train_s,
    epochs=50,
    batch_size = 40,
    validation_data=(X_test_s, y_test_s),
    verbose=10)
```

```python
import pickle

# Assuming you have trained a machine learning model and assigned it to the variable 'model'

# Save the model to a file
with open('skin_cancer.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
score = model.evaluate(X_test_s, y_test_s)
print('Test accuracy:', score)

```

```python
#plot the training and validation accuracy and loss at each epoch
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
